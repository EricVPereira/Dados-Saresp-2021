#Inep: Microdados de Alunos - SARESP - 2021

O Sistema de Avaliação de Rendimento Escolar do Estado de São Paulo (SARESP) é uma avaliação anual aplicada pela Secretaria da Educação do Estado de São Paulo com a finalidade de diagnosticar e acompanhar a evolução da educação básica paulista.
Os microdados de alunos do SARESP - 2021 apresentam os resultados das avaliações para cada aluno no ano de 2021. Acesse o Dicionário de microdados de alunos SARESP para a descrição dos campos do banco de dados.
